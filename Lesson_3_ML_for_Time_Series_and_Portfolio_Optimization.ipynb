## Introduction to Portfolio Optimization

Portfolio optimization is the process of selecting the best portfolio (asset allocation) out of the set of all feasible portfolios, according to some objective. The objective typically involves maximizing expected return for a given level of risk, or minimizing risk for a given level of expected return.

### Key Concepts in Portfolio Optimization

1.  **Risk**:
    *   **Definition**: In finance, risk refers to the uncertainty of an asset's future returns. It's the possibility that an investment's actual return will differ from its expected return.
    *   **Measurement**: Often quantified by the **standard deviation** of an asset's or portfolio's historical returns. A higher standard deviation indicates greater volatility and thus higher risk. Other measures include Value at Risk (VaR), Conditional Value at Risk (CVaR), and Beta (which measures systemic risk).

2.  **Return**:
    *   **Definition**: The gain or loss of an investment over a specified period, expressed as a percentage of the initial investment. It is the reward an investor receives for taking on risk.
    *   **Types**: Can be **historical return** (what has happened in the past) or **expected return** (the anticipated return an investment may generate over some future period). Expected return is typically estimated based on historical data, economic forecasts, and other relevant factors.

3.  **Diversification**:
    *   **Definition**: The strategy of investing in a variety of assets within a portfolio to minimize the impact of any single asset's performance on the overall portfolio. The core idea is "don't put all your eggs in one basket."
    *   **Role**: By combining assets that do not move in perfect lockstep (i.e., have low or negative correlation), diversification can reduce the overall risk of a portfolio without necessarily sacrificing expected returns. It primarily helps to mitigate unsystematic (specific) risk.

4.  **Efficient Frontier**:
    *   **Definition**: A concept introduced by Markowitz, the efficient frontier is a curve representing the set of optimal portfolios that offer the highest expected return for a given level of risk or the lowest risk for a given level of expected return. Any portfolio lying below the efficient frontier is suboptimal because it either offers less return for the same amount of risk or the same return for more risk.
    *   **Graphical Representation**: In a graph with portfolio risk (standard deviation) on the x-axis and portfolio expected return on the y-axis, the efficient frontier forms an upward-curving line. Investors aim to select a portfolio on this frontier that aligns with their individual risk tolerance.

### Markowitz Portfolio Theory (Modern Portfolio Theory - MPT)

*   **Foundational Concept**: Developed by Harry Markowitz in 1952, MPT is a framework for assembling a portfolio of assets such that the expected return is maximized for a given level of market risk. Alternatively, it aims to minimize risk for a given level of expected return.
*   **Main Idea**: MPT posits that investors can construct optimal portfolios by carefully choosing combinations of assets based on their individual risk, expected return, and, crucially, their **covariances (or correlations)** with other assets in the portfolio. Instead of evaluating individual investments in isolation, MPT emphasizes assessing how assets perform together within a portfolio.
*   **Diversification's Role**: A key tenet of MPT is that diversification among assets with imperfect correlations can lead to a portfolio with a lower overall risk than the weighted average risk of its individual components, for the same level of return. This is the benefit of combining assets that do not move in perfect unison.
*   **Assumptions**: MPT relies on several assumptions, such as investors being rational and risk-averse, having access to all information, and markets being efficient. While some assumptions are idealized, the core principles of MPT remain highly influential in investment management.

## Backtesting and Evaluation




### What is Backtesting?

Backtesting is a critical process in financial analysis that involves applying a trading strategy or analytical model to historical data to determine its accuracy and profitability. Essentially, it's a simulation of how a strategy would have performed if it had been implemented in the past.

**Purpose of Backtesting:**

1.  **Validation**: To confirm the effectiveness and robustness of an investment strategy before deploying it with real capital.
2.  **Performance Assessment**: To estimate the potential returns, risks, and other key performance metrics of a strategy under various market conditions.
3.  **Optimization**: To identify weaknesses and areas for improvement in a strategy, allowing for adjustments to enhance its risk-adjusted returns.
4.  **Risk Management**: To understand the potential downsides and drawdowns of a strategy, helping investors prepare for adverse market scenarios.

**General Methodology:**

The backtesting process typically involves the following steps:

1.  **Define the Strategy**: Clearly articulate the rules, indicators, and conditions that trigger buy, sell, or hold decisions.
2.  **Acquire Historical Data**: Obtain reliable historical market data (e.g., stock prices, trading volumes, economic indicators) relevant to the strategy.
3.  **Simulate Trades**: Apply the strategy rules to the historical data, step-by-step, to generate a series of simulated trades. This involves accounting for transaction costs (e.g., commissions, slippage) and market microstructure effects.
4.  **Calculate Performance Metrics**: Evaluate the simulated portfolio's performance using various metrics such as total return, volatility, Sharpe Ratio, Sortino Ratio, maximum drawdown, win rate, profit factor, etc.
5.  **Analyze Results**: Review the performance metrics to understand the strategy's strengths and weaknesses. Look for consistency, robustness, and any signs of overfitting.
6.  **Refine and Re-test**: Based on the analysis, modify the strategy as needed and repeat the backtesting process to see if improvements have been made.

It's crucial to use out-of-sample data (data not used in the strategy's development) for robust backtesting to avoid overfitting and ensure the strategy's real-world applicability.

### Evaluating Portfolio Performance: Key Metrics

After backtesting a strategy, evaluating its performance is crucial. This is done using several key metrics that provide insights into both the returns generated and the risks taken.

1.  **Sharpe Ratio**:
    *   **Definition**: The Sharpe Ratio measures the excess return (or risk premium) per unit of total risk in an investment or trading strategy. It is calculated as the portfolio's expected return minus the risk-free rate, divided by the portfolio's standard deviation of returns. (Risk-free rate is often assumed to be 0 for simplicity in many contexts, or a short-term government bond yield).
    *   **Formula**: $(R_p - R_f) / \sigma_p$, where $R_p$ is the portfolio return, $R_f$ is the risk-free rate, and $\sigma_p$ is the standard deviation of the portfolio's excess return.
    *   **Significance**: A higher Sharpe Ratio indicates a better risk-adjusted return. Investors prefer strategies with higher Sharpe Ratios because they imply that the portfolio is generating more return for the amount of risk taken. It helps to compare different investment opportunities on a consistent basis.

2.  **Sortino Ratio**:
    *   **Definition**: The Sortino Ratio is a modification of the Sharpe Ratio that differentiates harmful volatility from total volatility. It focuses only on the downside deviation (the standard deviation of negative returns), as investors are generally concerned with downside risk rather than upside volatility.
    *   **Formula**: $(R_p - R_f) / \sigma_{downside}$, where $R_p$ is the portfolio return, $R_f$ is the risk-free rate, and $\sigma_{downside}$ is the downside deviation.
    *   **Significance**: A higher Sortino Ratio indicates that the strategy is performing better relative to the undesirable (downside) volatility it experiences. It is often preferred by risk-averse investors because it only penalizes returns that fall below a specified target or the risk-free rate, providing a more accurate picture of a portfolio's risk-adjusted returns when downside protection is a key concern.

3.  **Maximum Drawdown (MDD)**:
    *   **Definition**: Maximum Drawdown is the largest percentage drop from a peak to a trough in the value of an investment portfolio over a specified period. It represents the worst historical loss an investor would have suffered if they had bought at the peak and sold at the bottom of the slump.
    *   **Calculation**: $(PeakValue - TroughValue) / PeakValue$.
    *   **Significance**: MDD is a crucial risk measure that quantifies the capital loss potential and the resilience of a strategy. It provides insight into the potential recovery time and the emotional impact an investment might have on an investor during adverse periods. A strategy with a lower maximum drawdown is generally preferred, as it indicates better capital preservation during market downturns. It's often considered alongside total return to balance potential gains with acceptable losses.

```python
import pandas as pd
import numpy as np

# Define a risk-free rate (can be adjusted)
risk_free_rate = 0.0

# Calculate annualized mean returns
annualized_returns = returns.mean() * 252

# Calculate annualized standard deviation (volatility)
annualized_volatility = returns.std() * np.sqrt(252)

# Calculate Sharpe Ratio
sharpe_ratios = (annualized_returns - risk_free_rate) / annualized_volatility

# Create a DataFrame to display the results
sharpe_df = pd.DataFrame({
    'Annualized Return': annualized_returns,
    'Annualized Volatility': annualized_volatility,
    'Sharpe Ratio': sharpe_ratios
})

print("Sharpe Ratios for each stock:")
print(sharpe_df.sort_values(by='Sharpe Ratio', ascending=False))
```

```python
num_portfolios = 10000 # Number of random portfolios to generate
results = np.zeros((3 + len(tickers), num_portfolios)) # Array to store results

for i in range(num_portfolios):
    weights = np.random.random(len(tickers)) # Generate random weights
    weights /= np.sum(weights) # Normalize weights to sum to 1

    # Calculate portfolio return
    portfolio_return = np.sum(returns.mean() * weights) * 252

    # Calculate portfolio volatility
    portfolio_std_dev = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))

    # Calculate Sharpe Ratio
    sharpe_ratio = (portfolio_return - risk_free_rate) / portfolio_std_dev

    # Store results
    results[0,i] = portfolio_return
    results[1,i] = portfolio_std_dev
    results[2,i] = sharpe_ratio
    for j in range(len(weights)):
        results[j+3,i] = weights[j]

# Create a DataFrame from the results
columns = ['Return', 'Volatility', 'Sharpe Ratio'] + [f'Weight_{s}' for s in tickers]
portfolio_df = pd.DataFrame(results.T, columns=columns)

# Find the portfolio with the maximum Sharpe Ratio
max_sharpe_portfolio = portfolio_df.loc[portfolio_df['Sharpe Ratio'].idxmax()]

print("Portfolio with Maximum Sharpe Ratio:")
print(max_sharpe_portfolio)
```

## Machine Learning for Portfolio Optimization

```python
import yfinance as yf
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
```

```python
print("Downloading data...")
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
# yfinance returns a MultiIndex DataFrame; we select 'Adj Close'
data = yf.download(tickers, start='2020-01-01', end='2026-01-01')['Close']

# Calculate daily returns and drop NaNs
returns = data.pct_change().dropna()
returns.head()
```

```python
num_assets = len(tickers)

# Create sliding windows
window_size = 10
returns_array = returns.values
X_list, y_list = [], []

for i in range(len(returns_array) - window_size):
    X_list.append(returns_array[i : i + window_size])
    y_list.append(returns_array[i + window_size])

X_np = np.array(X_list)
y_np = np.array(y_list)
print(X_np[0])
print(y_np[0])
```

```python
# Split into Training (80%) and Testing (20%) data chronologically
split_idx = int(len(X_np) * 0.8)
X_train, X_test = X_np[:split_idx], X_np[split_idx:]
y_train, y_test = y_np[:split_idx], y_np[split_idx:]

# Convert to PyTorch tensors and flatten X for the Feedforward Network
X_train_tensor = torch.tensor(X_train, dtype=torch.float32).view(len(X_train), -1)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32)

X_test_tensor = torch.tensor(X_test, dtype=torch.float32).view(len(X_test), -1)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32)
```

```python
# Create a DataLoader for batching during training
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

```

```python
class PortfolioNet(nn.Module):
    def __init__(self, num_features, num_assets):
        super(PortfolioNet, self).__init__()
        self.fc1 = nn.Linear(num_features, 64)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(64, num_assets)
        self.softmax = nn.Softmax(dim=-1) # Ensures weights are > 0 and sum to 1

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return self.softmax(x)
```

```python
def negative_sharpe_loss(weights, forward_returns, risk_free_rate=0.0):
    # Calculate portfolio returns for the batch
    portfolio_returns = torch.sum(weights * forward_returns, dim=1)

    mean_return = torch.mean(portfolio_returns)
    std_return = torch.std(portfolio_returns)

    # Calculate Sharpe (adding 1e-6 to avoid division by zero)
    sharpe_ratio = (mean_return - risk_free_rate) / (std_return + 1e-6)

    return -sharpe_ratio
```

```python
# Total features = 10 days * 5 assets = 50 features per sample
num_features = window_size * num_assets
model = PortfolioNet(num_features, num_assets)
optimizer = optim.Adam(model.parameters(), lr=0.001)

model
```

```python
epochs = 50
for epoch in range(epochs):
    model.train()
    epoch_loss = 0

    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()

        # 1. Model predicts weights
        predicted_weights = model(X_batch)

        # 2. Calculate loss based on actual next-day returns
        loss = negative_sharpe_loss(predicted_weights, y_batch)

        # 3. Backpropagate and update
        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        avg_loss = epoch_loss / len(train_loader)
        print(f"Epoch {epoch+1}/{epochs} | Avg Negative Sharpe Loss: {avg_loss:.4f}")
```

```python
model.eval() # Turn off training behaviors

with torch.no_grad():
    # Let's predict the weights for the very first day in our test set
    test_weights = model(X_test_tensor)

    first_day_weights = test_weights[0]

    print("\nPredicted Portfolio Allocation for the first test day:")
    for ticker, weight in zip(tickers, first_day_weights):
        print(f"{ticker}: {weight.item():.2%}")
```

## Time Series Data: Overview, Characteristics, Applications, and Key Concepts

### What is Time Series Data?
Time series data is a sequence of data points indexed (or listed) in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. Examples include daily stock prices, hourly temperature readings, or monthly sales figures. The unique aspect of time series data is its dependence on the order of observations, as each data point is not independent of previous or future data points.

### Key Characteristics of Time Series Data
1.  **Sequential Nature**: Observations are recorded in a specific temporal order, and this order is crucial for analysis.
2.  **Temporal Dependence**: The value of an observation at a given time point is often correlated with values at previous or subsequent time points. This is known as autocorrelation.
3.  **Irregularity**: While often collected at regular intervals, real-world time series can sometimes have missing values or irregular sampling.
4.  **Components**: Time series data can often be decomposed into several components: trend, seasonality, and residual (or noise).

### Common Applications of Time Series Analysis
Time series analysis is widely used across various fields for forecasting, understanding underlying causes, and anomaly detection. Some common applications include:
*   **Economics**: Forecasting GDP, inflation rates, unemployment rates, and consumer spending.
*   **Meteorology**: Predicting weather patterns, temperature, and rainfall.
*   **Engineering**: Monitoring sensor readings, system performance, and predictive maintenance.
*   **Retail**: Sales forecasting, inventory management, and demand planning.

#### Focus on Financial Applications
In finance, time series analysis is particularly critical due to the inherently temporal nature of financial markets. Key applications include:
*   **Stock Price Forecasting**: Predicting future stock prices, index movements (e.g., S&P 500), and cryptocurrency values to inform trading strategies and investment decisions.
*   **Economic Indicator Analysis**: Analyzing and forecasting economic indicators like interest rates, exchange rates, and commodity prices to assess market health and make policy decisions.
*   **Risk Management**: Modeling volatility (e.g., using GARCH models) and correlations between assets to manage portfolio risk.
*   **Algorithmic Trading**: Developing automated trading systems that react to patterns and predictions derived from time series data.
*   **Financial Auditing and Fraud Detection**: Identifying unusual patterns in transaction data that might indicate fraudulent activities.

### Key Concepts in Time Series Analysis
Understanding these concepts is fundamental to effective time series modeling:

1.  **Stationarity**:
    *   **Definition**: A time series is said to be stationary if its statistical properties (mean, variance, autocorrelation) do not change over time. In simpler terms, it looks the same no matter when you observe it.
    *   **Importance**: Many time series models (e.g., ARIMA) assume stationarity. Non-stationary series often need to be transformed (e.g., differencing) to achieve stationarity before modeling.
    *   **Types**: Strictly stationary (all statistical properties are constant) and weakly stationary (mean, variance, and autocovariance are constant).

2.  **Seasonality**:
    *   **Definition**: Seasonality refers to repetitive patterns or cycles that occur over a fixed period (e.g., daily, weekly, monthly, quarterly, yearly). These patterns are usually due to seasonal factors like time of year or day of the week.
    *   **Examples**: Retail sales peaking during holidays, higher electricity consumption in summer, or increased website traffic during weekdays.
    *   **Identification**: Often identified through visual inspection of plots, autocorrelation functions (ACF), or spectral analysis.

3.  **Trends**:
    *   **Definition**: A trend is a long-term increase or decrease in the data over time. It represents the underlying direction of the series, ignoring short-term fluctuations.
    *   **Types**: Can be linear (a constant rate of increase/decrease) or non-linear (e.g., exponential growth). It can also be upward, downward, or horizontal.
    *   **Examples**: A steady increase in global temperature over decades, a consistent growth in a company's revenue, or a long-term decline in a specific industry.
    *   **Importance**: Identifying and removing trends can reveal other components like seasonality and help in understanding the fundamental movement of the series.

### Additional Feature Engineering Techniques for Financial Time Series Data

Beyond basic normalization, differencing, and lag features, several other techniques can be employed to engineer features for predicting stock movements. These often involve technical indicators that summarize past price and volume data in a way that is believed to reflect market sentiment and momentum.

1.  **Moving Averages (MAs)**:
    *   **Simple Moving Average (SMA)**: Calculates the average price over a specified number of periods. For example, a 50-day SMA or 200-day SMA. SMAs help to smooth out price data and identify trends. Crossovers between different SMAs (e.g., 50-day SMA crossing above 200-day SMA) are often used as trading signals.
    *   **Exponential Moving Average (EMA)**: Similar to SMA but gives more weight to recent prices, making it more responsive to new information. EMAs are also used to identify trends and potential reversals.
    *   **Why useful**: MAs help in identifying trends, support, and resistance levels, and can generate buy/sell signals through crossovers.

### Other interesting Indicators

4.  **Bollinger Bands (BB)**:
    *   **Definition**: A volatility indicator consisting of a middle band (a simple moving average) and two outer bands (typically two standard deviations above and below the middle band). The bands expand and contract with market volatility.
    *   **Why useful**: Helps in identifying periods of high or low volatility, and potential price reversals when prices touch or break out of the bands.

5.  **Volume-based Indicators**:
    *   **On-Balance Volume (OBV)**: Relates volume to price change. If the closing price is higher than the previous close, all of the day's volume is considered up volume. If the closing price is lower, all of the day's volume is down volume.
    *   **Why useful**: Helps in confirming price trends. A rising OBV typically confirms an uptrend, while a falling OBV suggests a downtrend.

6.  **Volatility Measures**:
    *   **Standard Deviation of Returns**: A simple measure of historical volatility. Higher standard deviation indicates higher risk.
    *   **Average True Range (ATR)**: Measures market volatility by decomposing the entire range of an asset price for that period.
    *   **Why useful**: Volatility measures are crucial for risk management and understanding the expected range of price movements.

7.  **Date-based Features**:
    *   **Day of Week, Month, Year, Quarter**: These can capture cyclical patterns or seasonal effects in stock prices.
    *   **Why useful**: Helps models learn about potential seasonal or weekly biases in stock movements.

These engineered features can provide a richer set of information to a machine learning model, allowing it to better capture the complex dynamics of financial markets and potentially improve the accuracy of stock movement predictions.

```python
sequence_length = 50
# Function to create sequences for LSTM
def create_sequences(data, sequence_length):
    xs = []
    for i in range(len(data) - sequence_length + 1):
        x = data[i:(i + sequence_length)]
        xs.append(x)
    return np.array(xs)

# Restructure X_train_scaled and X_test_scaled into sequences
X_train_lstm = create_sequences(X_train, sequence_length)
X_test_lstm = create_sequences(X_test, sequence_length)

# Adjust y_train and y_test to align with the sequences
y_train_lstm = y_train[sequence_length - 1:] # Target corresponds to the last element of the sequence
y_test_lstm = y_test[sequence_length - 1:] # Target corresponds to the last element of the sequence

np.shape(X_train_lstm)
```

```python
import torch.nn as nn

# 1. Define the LSTM Model
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_layer_size=100, output_size=1, num_layers=1):
        super().__init__()
        self.hidden_layer_size = hidden_layer_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_layer_size, num_layers, batch_first=True)

        self.linear = nn.Linear(hidden_layer_size, output_size)

    def forward(self, input_seq):
        # Initialize hidden state and cell state
        h0 = torch.zeros(self.num_layers, input_seq.size(0), self.hidden_layer_size).to(input_seq.device)
        c0 = torch.zeros(self.num_layers, input_seq.size(0), self.hidden_layer_size).to(input_seq.device)

        lstm_out, _ = self.lstm(input_seq, (h0, c0))
        predictions = self.linear(lstm_out[:, -1, :])
        return predictions
```

```python
# 2. Instantiate the LSTMModel, Loss Function, and Optimizer
input_size = 1
hidden_layer_size = 50 # Using 50 as a suitable hidden_layer_size
output_size = 1
num_layers = 1
```

```python
model = LSTMModel(input_size=input_size, hidden_layer_size=hidden_layer_size, output_size=output_size, num_layers=num_layers)
loss_function = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
model
```
